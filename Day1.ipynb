```python
#Find two values in list whose sum is 0
a = [1,-1,5,7,2,9,4,3,2,-2]
#1.Brute force O(n^2)
#2.Time complexity O(n)
#3.Time complexity O(nlogn)
```

```python
#BRUTE FORCE
a = [1, -1, 5, 7, 2, 9, 4, 3, 2, -2]

def find_pairs_brute_force(arr):
    pairs = []
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == 0:
                pairs.append((arr[i], arr[j]))
    return pairs

pairs = find_pairs_brute_force(a)
pairs
```

```python
#HASH MAP(O(n^2))
#a = [1, -1, 5, 7, 2, 9, 4, 3, 2, -2]
def find_pairs_hash_map(arr):
    pairs = []
    seen = {}
    for num in arr:
        if -num in seen:
            pairs.append((num, -num))
        else:
            seen[num] = True
    return pairs

pairs_hash_map = find_pairs_hash_map(a)
pairs_hash_map
```

```python
#Using pointers O(nlogn)
def find_pairs_two_pointer(arr):
    pairs = []
    arr.sort()
    left = 0
    right = len(arr) - 1
    while left < right:
        current_sum = arr[left] + arr[right]
        if current_sum == 0:
            pairs.append((arr[left], arr[right]))
            left += 1
            right -= 1
        elif current_sum < 0:
            left += 1
        else:
            right -= 1
    return pairs
a = [1,-1,5,7,2,9,4,3,2,-2]
pairs_two_pointer = find_pairs_two_pointer(a)
pairs_two_pointer
```

```python
def subarray(arr):
    prefix_sum = 0
    prefix_sum_map = {}

    for i, num in enumerate(arr):
        prefix_sum += num

        # Check if prefix sum is zero
        if prefix_sum == 0:
            return arr[0:i+1]

        # Check if prefix sum is already seen
        if prefix_sum in prefix_sum_map:
            start_index = prefix_sum_map[prefix_sum] + 1
            return arr[start_index:i+1]

        # Store prefix sum with the current index
        prefix_sum_map[prefix_sum] = i

    return None

a = [1, -1, 5, 7, 2, 9, 4, 3, 2, -2]
result = subarray(a)
if result:
    print("Subarray with zero sum:", result)
else:
    print("No subarray with zero sum found")
```
