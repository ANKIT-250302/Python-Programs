#ASSUMPTIONS OF LINEAR REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("C:\\Users\\ajlad\\Downloads\\data.csv")
```

```python
df
```

```python
X = df.iloc[:,0:3].values
y = df.iloc[:,-1].values
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
# Residual
y_pred = model.predict(X_test)
residual = y_test - y_pred
```

##1.Linear relatopnship amoung Independent and dependent variables

```python
#Linearity check amoung dependent and independet vaiable
fig,(s1,s2,s3)=plt.subplots(ncols=3,figsize=(12,2.3))
s1.scatter(df['feature1'],df['target'])
s1.set_title("feature1")
s2.scatter(df['feature2'],df['target'])
s2.set_title("feature2")
s3.scatter(df['feature3'],df['target'])
s3.set_title("feature3")
```

##2.No Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = []

for i in range(X_train.shape[1]):
    vif.append(variance_inflation_factor(X_train, i))
```

```python
pd.DataFrame({'vif': vif}, index=df.columns[0:3]).T
#IF THE VALUE OF VIF IS GREATER THAN 5 IF SHOWS Multicollinearity
```

```python
# Another Technique
import seaborn as sns
sns.heatmap(df.iloc[:,0:3].corr(),annot=True)
```

##3.Normality of Residual

```python
sns.displot(residual,kind='kde')
```

```python
# QQ Plot

import scipy as sp

fig, ax = plt.subplots(figsize=(6,4))
sp.stats.probplot(residual, plot=ax, fit=True)

plt.show()
```

##4. Homoscedasticity

```python

plt.scatter(y_pred,residual)
```

##5. Autocorrelation of Residuals

```python
 plt.plot(residual)
```

##ENCODING THE DATA

```python
from sklearn.preprocessing import LabelEncoder
# Initialize LabelEncoder
label_encoder = LabelEncoder()


# Apply LabelEncoder to each column
df1['Extracurricular Activities'] = label_encoder.fit_transform(df1['Extracurricular Activities'])
```

##RESCALING THE DATASET

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
df_scaled = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)
```

```python
df_scaled
```

```python
x=df_scaled.iloc[:,0:5]
x
```

```python
y=df_scaled.iloc[:,-1]
y
```

#VIF-Variance Inflation Factor Calculation

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data['feature'] = x.columns
vif_data['VIF'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif_data=vif_data.sort_values(by='VIF', ascending=False)
print(vif_data)
```
