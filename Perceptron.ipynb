```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:\\Users\\ajlad\\Downloads\\Single Layer Perceptron Dataset.csv")
```

```python
data
```

```python
data=data.drop(columns=['Pattern'])
data=data.drop(13)
```

```python
print(data)
from sklearn.linear_model import Perceptron
p=Perceptron()
```

```python
x=data.iloc[:,1:3]
y = data.iloc[:, -1].astype(np.int_)
```

```python
x
```

```python
p.fit(x,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
# Define filler values for additional features
filler_feature_values = {i: x.iloc[:, i].mean() for i in range(x.shape[1]) if i not in [0, 1]}
filler_feature_ranges = {i: (x.iloc[:, i].min(), x.iloc[:, i].max()) for i in range(x.shape[1]) if i not in [0, 1]}
```

```python
# Plot decision regions
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values, y.values, clf=p, legend=2, 
                      filler_feature_values=filler_feature_values, 
                      filler_feature_ranges=filler_feature_ranges)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Decision Regions")
plt.show()
```
